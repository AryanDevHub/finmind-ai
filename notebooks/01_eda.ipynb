```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load data
print("Loading data...")
trans = pd.read_csv('../data/train_transaction.csv')
identity = pd.read_csv('../data/train_identity.csv')

print("✓ Data loaded successfully!")
print(f"\nTransaction dataset: {trans.shape[0]:,} rows × {trans.shape[1]} columns")
print(f"Identity dataset:    {identity.shape[0]:,} rows × {identity.shape[1]} columns")
```

```python
# Check fraud distribution
fraud_counts = trans['isFraud'].value_counts()
fraud_pct = trans['isFraud'].mean() * 100

print("=" * 40)
print("FRAUD DISTRIBUTION")
print("=" * 40)
print(f"Normal transactions: {fraud_counts[0]:,}  ({100 - fraud_pct:.1f}%)")
print(f"Fraud  transactions: {fraud_counts[1]:,}  ({fraud_pct:.1f}%)")
print(f"\nClass imbalance ratio: {fraud_counts[0]//fraud_counts[1]}:1")
print("\nThis is the problem we solve with SMOTE.")
```
